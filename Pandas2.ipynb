```python
# prime numbers
nums = [1, 2, 4, 5, 7, 9, 10]
# 2, 5, 7

def check_prime(num):
    if(num < 2): return False
    elif (num == 2): return True
    else:
        for i in range(2, num):
            if(num % i == 0): return False
        return True

prime_nos = []
for num in nums:
    if(check_prime(num)):
        prime_nos.append(num)

print(prime_nos)
```
