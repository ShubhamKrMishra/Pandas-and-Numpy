```python
df = pd.read_csv('./mtcars.csv')
# df.head()

# rename -> new dataframe
df = df.rename(columns={
    'Unnamed: 1': 'model',
    'mpg': 'mileage',
    'qsec': 'kmph'
})
# df

# sort by column names
columns_order = df.columns.sort_values()
df[columns_order]

# create new columns
# power -> hp * disp
df['power'] = df['hp'] * df['disp']
df['color'] = 'red'
df

# drop columns
df = df.drop(columns=['cyl', 'drat'])
df

# create a new file and store the data
# df.to_csv('new_cars.csv')
df.to_excel('new_cars1.xlsx')
```
